* getcwd: os에서 현재 directory를 받아오는 함수. 해당되는 파일을 열어라, 인코딩을 지정해줘야 한글이 깨지지않음. 그 파일을 이러한 인코딩으로 불러와라. 파일이름이랑 인코딩만 바꾸어주면 현재 directory에 있는 text파일을 불러오게 된다. 이 text가 raw에 들어오게 됨.

```python
raw = open(os.getcwd()+r'/06_01.txt', encoding = 'utf8').read()
```

```python
print(type(raw))
print(len(raw))
print(raw[:160])
```

* nltk패키지 중 word_tokernize라는 함수를 사용해서 raw값을 처리하고 이것을 tokens라는 변수에 넣는다.
* word_tokernize는 띄어쓰기는 제외하고 단어만 뽑아주는 함수임, 단어의 중복도 발생 앞에서부터 쭉 가져옴

```python
tokens = nltk.word_tokenize(raw)
print(type(tokens))
print(len(tokens))
print(tokens[:30])
```

* nltk.Text라는 함수를 사용해서 nltk 전용 variable로 변환. type은 nltk.text.Text임.
* 변환한 tokens를 text변수에 넣어주는데, collocations라는 함수를 사용하려면 nltk.text.Text로 변환해줘야함.
* collocations함수는 같이 나올 확률이 높은 두개의 단어를 보여준다. 대부분 등장 인물의 이름이다.

```python
text = nltk.Text(tokens)
print(type(text))
print(text[:10])
text.collocations()
```

* text 속에서 great가 들어있는 모든 문장을 찾아내어라. 문맥을 보고 싶으니까 great라는 단어를 중심으로 양쪽으로 총 79개의 character를 보면 이 'great'가 문맥 속에서 어떻게 사용되는지 알 수 있는 것임. 79개 character, 처음부터 25번째 great까지 뽑아낸 것.

```python
text.concordance('great', 79, 25)
```

```python
text.concordance('woman', 100, 10)
```

1 모비딕.txt를 raw라는 변수에 넣는다

2 200번째 string까지 프린트한다

3 FreqDist함수를 사용해서 raw에서 가장 빈번하게 나온 단어를 fdist에 넣는다

4 most_common()함수를 사용해서 가장 빈번하게 나온, 상위 50개 단어를 프린트한다

* ch.lower() list를 만들건데 raw 안에서 ch에 해당하는, 각 character가 알파벳이면 ch에 넣어서 그것을 소문자로 list에 넣어준다

5 fdist에 ch.lower() 리스트에 FreqDist한 값을 넣는다

6 fdist중 가장 빈번하게 나온 상위 5개 character를 프린트한다

7 fdist.most_common(5)은 두개의 정보가 각각(character, count) 돌아가는 리스트. 이 리스트에서 char만 가져와라

```python
raw = nltk.corpus.gutenberg.raw('melville-moby_dick.txt')
print(raw[:200])
fdist = nltk.FreqDist(raw)
print(fdist.most_common(50))


(ch.lower() for ch in raw if ch.isalpha())


fdist = nltk.FreqDist(ch.lower() for ch in raw if ch.isalpha())
print(fdist.most_common(5))
[char for (char, count) in fdist.most_common(5)]
```
