```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
html = urlopen("http://www.ticketlink.co.kr/ranking/reserve/daily?categoryId=10")
html
```

```python
bsObject = BeautifulSoup(html, "html.parser")
bsObject
```

```python
bsObject.tbody.select('tr')[0]
```

```python
for n in range(0, 10):
    clue = bsObject.tbody.select('tr')[n]
    title = clue.find("td",{"class":"tl tx"}).get_text() 
    rate = clue.find("td",{"class":"ff_tahoma"}).get_text()
    duration = clue.find("td",{"class":"tl ff_tahoma"}).get_text()
    location = clue.find("td",{"class":"tl desc_vt"}).get_text()
    print(n+1, title, rate, duration, location)
  
```

```python
with open('ticketlink.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['rank', 'title', 'rate', 'duration', 'location'])
    
    for n in range(0, 10):
        clue = bsObject.tbody.select('tr')[n]
        title = clue.find("td",{"class":"tl tx"}).get_text() 
        rate = clue.find("td",{"class":"ff_tahoma"}).get_text()
        duration = clue.find("td",{"class":"tl ff_tahoma"}).get_text()
        location = clue.find("td",{"class":"tl desc_vt"}).get_text()
        writer.writerow([n+1, title, rate, duration, location])
  
```
